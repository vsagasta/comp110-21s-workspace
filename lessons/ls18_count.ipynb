# Quantifying Shakespeare

Read the complete works of shakespeare and count every letter

```python
lines: list[str] = []
file_handle = open("../data/shakespeare.txt", "r", encoding="utf8")

for line in file_handle:
    line_lowercase: str = line.lower()
    lines.append(line_lowercase)

file_handle.close()
print(f"{len(lines)} lines of text read")
```

```python
print(lines[0])
```

## Count the characters shakespeare used

Use a dictionary, where each key is a letter and each value will be the number of times we encounter that character when processing the lines we just read in.

```python
counts: dict[str, int] = {}

for line in lines:
    for character in line:
        if character.isalpha():
            if character in counts: # is character a key in the counts dictionary
                counts[character] += 1
            else:
                counts[character] = 1
print(counts)

```

```python
from matplotlib import pyplot as plt

figure, axes = plt.subplots(1)
axes.bar(counts.keys(), counts.values())
```

```python
sorted_counts: dict[str, int] = dict(sorted(counts.items()))

figure, axes = plt.subplots()
axes.bar(sorted_counts.keys(), sorted_counts.values())
```
